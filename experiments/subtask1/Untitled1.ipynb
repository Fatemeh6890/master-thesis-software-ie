# Bert-base-uncased negative sampling ratio

```python
import pandas as pd
import ast
import json
```

```python
with open("../../evaluation/subtask1/subtask1_bert_base_uncased_negative_sampling_results.txt") as f:
    prediction_infos = f.readlines()
print(prediction_infos)
```

```python
result_of_predictions = []
for pred in prediction_infos:
    result_of_predictions.append(ast.literal_eval(pred))
```

```python
pd.DataFrame(result_of_predictions)
```

```python
metrics = []
for d in result_of_predictions:
    metrics.append({
        'name': d['negative_sampling_ratio'],
        'precision': d['precision'],
        'recall': d['recall'],
        'f1': d['f1']
    })
```

```python
# Converting to DataFrame
df = pd.DataFrame(metrics)
df.head()
```

```python
print(df.to_latex())
```

# Scibert negative sampling ratio

```python
with open("../../evaluation/subtask1/subtask1_allenai_scibert_scivocab_uncased_negative_sampling_results.txt") as f:
    bert_base_uncased_negative = f.readlines()
result_of_predictions_bert_base_uncased_negative = []
for pred in bert_base_uncased_negative:
    result_of_predictions_bert_base_uncased_negative.append(ast.literal_eval(pred))
    
metrics_bert_base_uncased_negative = []
for d in result_of_predictions_bert_base_uncased_negative:
    metrics_bert_base_uncased_negative.append({
        'negative_sampling_ration': d['negative_sampling_ratio'],
        'precision': d['precision'],
        'recall': d['recall'],
        'f1': d['f1']
    })


# # Converting to DataFrame
df_bet_bae_uncased_negative = pd.DataFrame(metrics_bert_base_uncased_negative)
df_bet_bae_uncased_negative.head(6)
```
